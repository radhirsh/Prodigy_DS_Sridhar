```python
#Before going to visualization data cleaning is must
df=df.dropna()
df.isnull().sum()
```

# Data Visualization part

```python
import matplotlib.pyplot as plt
df1=df[["Country Name", "2022"]].sort_values(by="2022",ascending=False).head(50)
plt.figure(figsize=(15,10))
bars=plt.bar(x=df1["Country Name"],height=df1["2022"],align="center")
plt.xlabel("Country Name")
plt.ylabel("Population")
plt.title("Top 50 countries Population distribution in 2022")
plt.xticks(rotation=90)

plt.show()


```

```python
plt.hist(x=df1['2022'], bins=10, orientation="vertical", color='red', stacked=True, label="2022 Population")

plt.xlabel("Population in 2022")
plt.ylabel("Frequency")
plt.title("Histogram of 2022 Population")

plt.legend()

plt.show()
```

# Enhancing visualization clarity with Seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns

s = df[["Country Name", "2022"]].sort_values(by="2022", ascending=False).head(50)

# Create the horizontal bar plot using seaborn
plt.figure(figsize=(10, 10))
sns.barplot(y=s["Country Name"], x=s['2022'], orient="h")

# Rotate x-axis labels
plt.xticks(rotation=0)
plt.title("Distribution of Top 50 countries in 2022")
plt.legend()
# Show the plot
plt.show()

```

# Data Encoding

There are some models like Linear Regression which does not work with categorical dataset in that case we should try to encode categorical dataset into the numerical column ,But here we dont want to use data encoding because all variables are consistent

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15, 6))
df1=df[["Country Code","2022"]].sort_values(by="2022",ascending=False).head(50)
sns.scatterplot( x="Country Code", y='2022', data=df1,hue='Country Code', size='2022')
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.xticks(rotation=70,fontsize=10)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15, 6))
df1=df[["2012","2022"]]
sns.scatterplot( x="2012", y='2022' ,data=df1, size='2022',label='Population')
plt.title('Population in 2012 vs 2022')
plt.xlabel('Population in 2012')
plt.ylabel('Population in 2022')
x = df1["2012"]
y = df1["2022"]
coefficients = np.polyfit(x, y, 1)
m, b = coefficients

# Plot the linear regression line
plt.plot(x, m*x + b, color='red', linestyle='--', label='Regression Line')
plt.legend(title="Size", loc="upper left", fontsize="small", markerscale=1)


plt.show()
```
